##Importar bibliotecas

```python
import pandas as pd
import seaborn as sns

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=200,
                           n_features=10,
                           n_informative=2,
                           n_redundant=6,
                           n_classes=2,
                           n_clusters_per_class=1,
                           n_repeated=2,
                           class_sep=0.9)
```

##Leitura do conjunto de dados

```python
df = pd.DataFrame(X)
df["Rotulo"] = y.tolist()
```

##Primeiras an√°lises do conjunto de dados

```python
sns.histplot(df, x=0, hue="Rotulo", kde=True)
```

```python
sns.pairplot(df, hue="Rotulo")
```

## Separa√ß√£o do conjunto de dados em treino e teste

```python
X= df.values[:,:-1]
y= df.values[:,-1]

xTrain, xTest, yTrain, yTest = train_test_split(X,y, test_size=0.3)
```

##Inicializa√ß√£o e execu√ß√£o dos algoritmos de aprendizado

```python
modeloRegressao = LogisticRegression()
modeloArvore = DecisionTreeClassifier()
modeloFloresta= RandomForestClassifier()
```

```python
#inducao dos algoritmos de aprendizado

modeloRegressao.fit(xTrain,yTrain)
modeloArvore.fit(xTrain, yTrain)
modeloFloresta.fit(xTrain, yTrain)
```

##Predi√ß√£o no conjunto de teste

```python
yRegressao = modeloRegressao.predict(xTest)
yArvore = modeloArvore.predict(xTest)
yFloresta= modeloFloresta.predict(xTest)
```

```python
resultado = pd.DataFrame()

resultado["RotuloOriginal"] = yTest.tolist()
resultado["Regressao"] = yRegressao.tolist()
resultado["Arvore"] = yArvore.tolist()
resultado["Floresta"] = yFloresta.tolist()
```
